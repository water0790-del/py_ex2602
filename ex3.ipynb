```python
#인스턴스 속성 사용하기
class Person():
    def __init__(self):
        self.hello = "안녕하세요" #인스턴스 속성

    def hi(self):
        print(self.hello)

jin = Person()  # 이 때 __init__ 메서드는 이미 실행됐다.
jin.hi()
```

```python
#클래스(객체를 표현하기 위한 문법의 일종이다.) 
#프로그래밍으로 객체를 만들 때 사용하는 것
#- 실체가 다른 어떤 실체를 바라볼때 객체라고 함??
#게임 케리터도 객체, 아이템도 객체
#전사, 마법사, 궁수, 자동차, 비행기 -> object
# 현실, 가상, 개념이 될 수 있다.
# 객체를 만들때 두가지(속성, 능력)
#속성(명사적)*저장하는것 : attribute
#능력(기능/ 동사적)*실행하는것 : method (메쏘드)


class person:   #설계도(붕어빵틀): 인스턴스를 만들기 위한 도구?
     def hello(self):
        print("hi")

jin = person() # jin은 person의 인스턴스(실제객체)  #jin은 인스턴스변수
jin.hello()
```

```python
class Person:
  folder = [] # 클래스 속성(공공제) -> 클래스에 속한다. 모든 인스턴스에서 접근이 가능하다.

  def add_files(self, file):
    Person.folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```
