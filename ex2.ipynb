```python
print("예제 1")
s = "안녕하세요"

print("s길이 출력")
print(len(s))

#s[start:end(-1):step]
print("s를 이루는 문자 중에서 앞에 거 2개 출력")
print(s[0:2])

print("s를 이루는 문자 중에서 문자하나 건너 뛰고 1개 출력")
print(s[1:2])

print("s를 이루는 문자 중에서 문자 4개 출력, 첫 인덱스를 생략")
print(s[:4])

print("s를 이루는 문자 중에서 앞에 3개 건너뛰고 나머지 출력,len으로")
print(s[3:len(s)])

print("s를 이루는 문자 중에서 앞에 3개 건너뛰고 나머지 출력,l: 뒤에 안적어도 같음")
print(s[3:])

print("예제 2")
s = "안녕하세요"

print("S를 이루는 문자르 하나씩 출력, while사용")
i = len(s) - 1
while i >=0:
    print(s[i])
    i -= 1

print("S를 이루는 문자르 하나씩 출력, for사용")
for c in s:
    print(c)
```
