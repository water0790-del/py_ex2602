```python
import matplotlib.pyplot as plt

fig = plt.figure() # 1. Figure(대문자)를 figure(소문자)로 수정

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]

#x축 라벨 이름(tick이름)
labels = ["english", "math"]
colors = ["red", "blue"]

# 2. barh로 유지 (가로 막대)
# 3. tick_lavels를 tick_label로 수정 (오타 및 철자 수정)
a.barh(x, y, tick_label=labels, color=colors)           

plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure() # 1. Figure(대문자)를 figure(소문자)로 수정

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]

#x축 라벨 이름(tick이름)
labels = ["english", "math"]
colors = ["red", "blue"]

# 2. barh로 유지 (가로 막대)
# 3. tick_lavels를 tick_label로 수정 (오타 및 철자 수정)
a.barh(x, y, tick_label=labels, color=colors)      

for i, v in enumerate(y):
    a.text(v+2, i + 1.05, str(v))


plt.show()
```
