```python
from openpyxl import *

# ==== 딕셔너리로 값 생성 후 엑셀에 넣기 ====

item1 = {
    '품목': '바나나',
    '가격': '3000',
    '수량': 1,
    '날짜': '2026년 2월 20일'
}

item2 = {
    '품목': '책',
    '가격': '15000',
    '수량': 3,
    '날짜': '2026년 2월 22일'
}

item3 = {
    '품목': '커피',
    '가격': '4000',
    '수량': 10,
    '날짜': '2026년 2월 23일'
}

item4 = {
    '품목': '신발',
    '가격': '50000',
    '수량': 4,
    '날짜': '2026년 2월 26일'
}

wb = Workbook()
ws = wb.active

ws['A1'] = '품목' # A열 1행 : 품목
ws['B1'] = '가격' # B열 1행 : 품목
ws['C1'] = '수량' # C열 1행 : 품목
ws['D1'] = '날짜' # D열 1행 : 품목

items = [item1, item2, item3, item4]

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1

wb.save('test2.xlsx')
```
